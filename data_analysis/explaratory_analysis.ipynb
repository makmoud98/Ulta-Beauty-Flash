# Preprocessing Data

```python
#Drop Duplicated and Drop useless columns
product_catalog = product_catalog.drop(['IS_DEFAULT_SKU','ONLINE_ONLY', 'DESCRIPTION'], axis=1)
```

```python
product_catalog = product_catalog.drop_duplicates(subset = ["SKU_ID"])
product_catalog = pd.merge(left=product_catalog, right=sku_meta_data, left_on='SKU_ID', right_on='SKU_ID')
```

```python
unique_brands = product_catalog['BRAND_NAME'].unique()
category_name = product_catalog['CATEGORY_NAME'].unique()
```

```python
columns = np.concatenate([unique_brands, category_name])
```

```python
# product_catalog.loc[product_catalog['SKU_ID'] == 1031082]['BRAND_NAME'].values[0]
```

```python
np.where(unique_brands == 'Ardell')
```

```python
#Turning the dataset into a item matrix
brand_vectors = []
for curid in product_catalog['SKU_ID'].unique():

    zeros = np.zeros(len(columns))
    brand = product_catalog.loc[product_catalog['SKU_ID'] == curid]['BRAND_NAME'].values[0]
    category = product_catalog.loc[product_catalog['SKU_ID'] == curid]['CATEGORY_NAME'].values[0]
    
    brand_index = np.where(unique_brands == brand)[0][0]
    category_index = np.where(category_name == category)[0][0]
    
    zeros[brand_index] = 1
    zeros[category_index + len(unique_brands) - 1] = 1
    brand_vectors.append(zeros)
    
```

```python
brand_matrix = pd.DataFrame(data=brand_vectors, index=product_catalog['SKU_ID'].unique(), columns=columns)
```

```python
brand_matrix.index.name = 'SKU_ID'
```

```python
brand_matrix = pd.merge(left=brand_matrix, right=sku_meta_data, left_on='SKU_ID', right_on='SKU_ID')
```

# Mocking Data using Statistical Methods

```python
#loading the user matrix
user_pd = pd.read_csv('user_data.csv')
```

```python
user_pd.head()
```

```python
unique_sku = product_catalog['SKU_ID'].unique()
user_id = user_pd['id'].unique()
```

```python
#creating the utility matrix using mock data
samples = np.random.normal(1, 0.1, len(unique_sku))
user_recommendations = []
for user in user_id:
    data = np.zeros(len(unique_sku))
    for index, sku in enumerate(unique_sku): 
        turn = np.random.uniform(0,2)
        multiplier = random.randint(0, 10)
        if turn <= samples[index]:
            data[index] = 1 * multiplier
    user_recommendations.append(data)
```

```python
def find_top_n(lst, n):
    lst = [x[-1] for x in lst] # remove tups 
    choices = np.random.choice(lst, n)
    # remove dups 
    choices = set(choices) 
    choices = list(choices)
    return choices
```

```python
def find_keyword(keyword):
    index = category_array.index(keyword)
    return lst_recs[index]
```
